```python
import requests
url="https://www.accuweather.com/en/in/bengaluru/204108/daily-weather-forecast/204108"

response=requests.get(url,headers={'user-agent':'hssh'})
response.content
```

```python
from bs4 import BeautifulSoup

soup=BeautifulSoup(response.content,'html.parser')
print(soup.prettify())
```

```python
date=soup.find_all('span',class_='module-header sub date')

date=[i.text for i in date]
date
```

```python
high_tem=soup.find_all('span',class_='high')

high_tem=[i.text for i in high_tem]

high_tem
```

```python
low_tem=soup.find_all('span',class_='low')

low_tem=[i.text for i in low_tem]

low_tem
```

```python
waterdrop=soup.find_all('div',class_='precip')

waterdrop=[i.text for i in waterdrop]

waterdrop
```

```python
import pandas as pd

df=pd.DataFrame({'date':date,'high_tem':high_tem,'low_tem':low_tem,'waterdrop':waterdrop})

df
```

```python
import pandas as pd
df=pd.DataFrame({'date':date,'high_tem':high_tem,'low_tem':low_tem,'waterdrop':waterdrop})
df['high_tem']=pd.to_numeric(df['high_tem'].str.replace('',' '))
df['low_tem']=pd.to_numeric(df['low_tem'].str.replace('',' ').str.replace('/',' '))
df['waterdrop']=pd.to_numeric(df['waterdrop'].str.replace('\n\n\t\t',' ').str.replace('%\n\t',' '))
df
```

```python
df["date"] = df["date"] + "/2023"
```

```python
df['date']= pd.to_datetime(df['date'],format='%m/%d/%Y', errors='coerce')
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

ax1=plt.subplot()
ax2=ax1.twinx()

ax1.bar(df['date'],df['waterdrop'],color='y')
ax1.set_xlabel('date')
ax1.set_ylabel('waterdrop(%)')

ax2.plot(df['date'],df['high_tem'],color='r',marker='o')
ax2.plot(df['date'],df['low_tem'],color='g',marker='x')

ax2.set_ylabel('Temperature F')
```
